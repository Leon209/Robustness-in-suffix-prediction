```python
# Cell 10: Function to process any dataset with saved encoder
def process_dataset_with_saved_encoder(csv_path, result_name, encoder_path, properties_path):
    """
    Process any CSV dataset using a pre-saved encoder to ensure consistent encoding.
    
    Args:
        csv_path: Path to the CSV file to process
        result_name: Name for the output files
        encoder_path: Path to the saved encoder_decoder.pkl
        properties_path: Path to the saved event_log_properties.pkl
    """
    import pickle
    import torch
    from event_log_loader.new_event_log_loader import CSV2EventLog, EventLogDataset, EventLogSplitter
    
    # Load the saved encoder and properties
    encoder_decoder = torch.load(encoder_path, weights_only=False)
    with open(properties_path, 'rb') as f:
        event_log_properties = pickle.load(f)
    
    print(f"ðŸ”„ Processing dataset: {csv_path}")
    print(f"ðŸ“ Using encoder from: {encoder_path}")
    
    # Create event log from CSV
    event_log = CSV2EventLog(csv_path, **event_log_properties)
    
    # Process train, val, test splits
    splitter = EventLogSplitter(**event_log_properties)
    train_df, val_df, test_df = splitter.split(event_log)
    
    # Encode each split using the saved encoder
    # print("ðŸ“Š Encoding train dataset...")
    # train_encoded, train_categories = encoder_decoder.encode_df(train_df)
    # train_dataset = EventLogDataset(train_encoded, train_categories, encoder_decoder)
    # torch.save(train_dataset, f'../../../encoded_data/{result_name}_{encoder_decoder.min_suffix_size}_train.pkl')
    
    # print("ðŸ“Š Encoding validation dataset...")
    # val_encoded, val_categories = encoder_decoder.encode_df(val_df)
    # val_dataset = EventLogDataset(val_encoded, val_categories, encoder_decoder)
    # torch.save(val_dataset, f'../../../encoded_data/{result_name}_{encoder_decoder.min_suffix_size}_val.pkl')
    
    print("Encoding test dataset...")
    test_encoded, test_categories = encoder_decoder.encode_df(test_df)
    test_dataset = EventLogDataset(test_encoded, test_categories, encoder_decoder)
    torch.save(test_dataset, f'../../../encoded_data/{result_name}_test.pkl')
    
    print(f"All datasets saved with prefix: {result_name}")

    

print("Function 'process_dataset_with_saved_encoder' defined successfully!")
```

```python
# Cell 11: Process perturbed dataset with consistent encoding
print("ðŸš€ Processing perturbed dataset with consistent encoding...")

process_dataset_with_saved_encoder(
    csv_path='../../../data/helpdesk_large_perturbations.csv',
    result_name='helpdesk_large_perturbations',
    encoder_path='../../../encoded_data/data_encoder/helpdesk_encoder_decoder.pkl',
    properties_path='../../../encoded_data/data_encoder/helpdesk_event_log_properties.pkl'
)

print("ðŸŽ‰ Perturbed dataset processing completed!")
```
